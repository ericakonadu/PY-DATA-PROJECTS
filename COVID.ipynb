```python
import pandas as pd
import numpy as np

import matplotlib.pyplot as plt
import matplotlib as mpl
import seaborn as sns
%matplotlib inline
mpl.style.use("seaborn")



df=pd.read_excel('PA_singlestate_timeseries.xlsx')
```

### ACTUAL CASES 
From the graph below, overtime, covid cases rose and reached a peak around 2020-11 and 2020-12. Then the cases gradually began to reduce.

```python
x=df['date']
y1=df['actuals.newCases']
# y2=df['actuals.vaccinationsCompleted']


plt.plot(x,y1)
# plt.plot(x,y2)

plt.title('COVID CASES OVER TIME')
plt.xlabel('NUMBER OF CASES')
plt.ylabel('DATE')
plt.show()
```

### VACCINATION
From the start of 2021, vaccines were introduced and they were distributed to a lot more people over time. 

```python
x=df['date']
y2=df['actuals.vaccinationsCompleted']

plt.plot(x,y2)
plt.title('VACCINATIONS OVER TIME')
plt.xlabel('DATE')
plt.ylabel('NUMBER OF VACCINES')
plt.show()
```

### RELATIONSHIP BETWEEN COVID CASES AND VACCINATIONS
From the graph below, it shows when the covid vaccines were administered, the number of covid cases reduced immensely.

```python
x=df['actuals.newCases']
y=df['actuals.vaccinationsCompleted']


plt.plot(x,y)
plt.title('COVID CASES AND VACCINATIONS COMPLETED')
plt.xlabel('COVID CASES')
plt.ylabel('VACCINATIONS COMPLETED')
plt.show()
```
