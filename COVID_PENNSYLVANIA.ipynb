```python
import pandas as pd
import numpy as np

import matplotlib.pyplot as plt
import matplotlib as mpl
import seaborn as sns
%matplotlib inline
mpl.style.use("seaborn")



df=pd.read_excel('PA_singlestate_timeseries.xlsx')
```

### ACTUAL CASES 
From the graph below, overtime, covid cases rose and reached a peak around 2020-11 and 2020-12. Then the cases gradually began to reduce.

```python
x=df['date']
y1=df['actuals.newCases']

plt.plot(x,y1, c= '#008B8B', linewidth = '2.0')

font1 = {'family':'serif','color':'black','size':15}
font2 = {'family':'serif','color':'black','size':12}

plt.title('COVID CASES OVER TIME', fontdict = font1)
plt.xlabel('NUMBER OF CASES',fontdict = font2)
plt.ylabel('DATE', fontdict = font2)
plt.show()
```

### VACCINATION
From the start of 2021, vaccines were introduced and they were distributed to a lot more people over time. 

```python
x=df['date']
y2=df['actuals.vaccinationsCompleted']

plt.plot(x,y2, linestyle = 'dashdot', c= 'r', linewidth = '3.0')
plt.title('VACCINATIONS OVER TIME', fontdict = font1)
plt.xlabel('DATE', fontdict = font2)
plt.ylabel('NUMBER OF VACCINES', fontdict = font2)
plt.show()
```

### RELATIONSHIP BETWEEN COVID CASES AND VACCINATIONS
From the graph below, it shows when the covid vaccines were administered, the number of covid cases reduced immensely.

```python
x=df['actuals.newCases']
y=df['actuals.vaccinationsCompleted']


plt.plot(x,y, c='#C71585')
plt.title('COVID CASES AND VACCINATIONS COMPLETED', fontdict = font1)
plt.xlabel('COVID CASES', fontdict = font2)
plt.ylabel('VACCINATIONS COMPLETED', fontdict = font2)
plt.show()
```
